subplot()函数将画布分区,scatter绘制散点图

```python
#正弦曲线和余弦曲线，subplot，plot，scatter
import numpy as np

#生成x的坐标（0-10的100个等差数列）
x = np.linspace(0,10,100)
sin_x = np.sin(x)
cos_x = np.cos(x)

plt.subplot(2,2,1)#将画布分为2行2列，将图画到画布的1区域
plt.plot(x,sin_x)#绘制正弦曲线

plt.subplot(2,2,3)#将画布分为2行2列，将图画到画布的3区域
plt.plot(x,cos_x)#绘制余弦曲线

plt.subplot(2,2,2)#将画布分为2行2列，将图画到画布的2区域
plt.scatter(x,sin_x)#绘制散点图

plt.show()
```

```python
#使用scatter画10中大小100中颜色的散点图
# np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
size = np.random.rand(100)*1000
plt.scatter(x,y,c=colors,s=size,alpha=0.7)
plt.show()
```

bar函数可以绘制柱状图，barh将y和x轴对换 竖着方向为x轴
bar.set设置柱状图的颜色等,plt.legend()添加图例

```python
#使用bar绘制柱状图，并设置柱的宽度
x=[1980,1985,1990,1995]
x_labels=['1980年','1985年','1990年','1995年']
y=[1000,2000,3000,4000]
plt.bar(x,y,width=3)

plt.xticks(x,x_labels)
plt.xlabel('年份')
plt.ylabel('销量')
plt.title('年份销量对比图')

plt.show()
```

```python
#使用bar和barh绘制柱状图
x=np.arange(5)
y=np.random.randint(-5,5,5)


plt.subplot(1,2,1)
v_bar=plt.bar(x,y,color='lightblue')
#在第一列的画布中 0位置画一条蓝线
plt.axhline(0,color='blue',linewidth=2)
for bar,height in zip(v_bar,y):
    if height<0:
        bar.set(edgecolor='darkred',color='lightgreen',linewidth='3')

#barh将y和x轴对换 竖着方向为x轴
plt.subplot(1,2,2)
plt.barh(x,y)
#在第二列的画布中0位置处画蓝色的线
plt.axvline(0,color='red',linewidth=2)

plt.show()
```

```python
#柱状图使用实例
#三天中三部电影的票房变化
real_names=['千与千寻','玩具总动员4','黑衣人：全球追缉']
real_num1=[5453,7548,6543]
real_num2=[1840,4013,3421]
real_num3=[1080,1673,2342]

#生成x  第1天   第2天   第3天
x =np.arange(len(real_names))
x_label=['第{}天'.format(i+1) for i in range(len(real_names))]

#绘制柱状图
#设置柱的宽度
width=0.3
plt.bar(x,real_num1,color='g',width=width,label=real_names[0])
plt.bar([i+width for i in x],real_num2,color='b',width=width,label=real_names[1])
plt.bar([i+width*2 for i in x],real_num3,color='r',width=width,label=real_names[2])

#修改x坐标
plt.xticks([i+width for i in x],x_label)
#添加图例
plt.legend()
#添加标题
plt.title('3天的票房数')

plt.show()
```

pie函数可以绘制饼状图

```python
#饼图的绘制
languages = ['java','html','sql']
popularity = [59210,55466,45677]

#百分比autopct,顺时针counterclock，脱离startangle
plt.pie(popularity,labels=languages,autopct='%1.1f%%',counterclock=False,startangle=90,explode=[0,0,0.1])
plt.tight_layout()
plt.show()
```

```python
#使用normal函数生成1000个正太分布的随机数，使用hist函数绘制这100个随机数的分布状态
#几个直方图画到一个画布中,第一个参数期望  第二个均值
x1=np.random.normal(0,0.8,1000)
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(3,2,1000)
#参数分别是bins：装箱，alpha：透明度
kwargs=dict(bins=100,alpha=0.4)
plt.hist(x1,**kwargs)
plt.hist(x2,**kwargs)
plt.hist(x3,**kwargs)
plt.show()
```

使用contourf绘制等高线图

```python
#使用pyplot绘制等高线图

x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
#计算x和y的相交点a
X,Y=np.meshgrid(x,y)
# 计算Z的坐标
Z=np.sqrt(X**2+Y**2)
plt.contourf(X,Y,Z)
# plt.contour(X,Y,Z)
# 颜色越深表示值越小，中间的黑色表示z=0.

plt.show()
```

通过 reshape 方法可以将一维数组变成二维、三维或者多维数组。

通过 ravel 方法或 flatten方法可以将多维数组变成一维数组。

改变数组的维度还可以直接设置 Numpy 数组的 shape属性（元组类型），通过 resize 方法也可以改变数组的维度。

```python
#reshape，ravel，flatten
a=np.arange(24)
print(a)
print('数组 a 的维度：',a.shape)
print('-'*30)

#使用 reshape 将一维数组变成三维数组
b=a.reshape(2,3,4)
print(b)
print('数组 b 的维度：',b.shape)
print('-'*30)

#将 a 变成二维数组
c=a.reshape(3,8)
print(c)
print('数组 c 的维度：',c.shape)
print('-'*30)

#使用 ravel 函数将三维的 b 变成一维的数组
a1=b.ravel()
print(a1)
print('-'*30)

#使用 flatten 函数将二维的 c 变成一维的数组
a2=c.flatten()
print(a2)
print('-'*30)
```

```python
数组的拼接：concatenate，hstack，vstack
```

```python
#concatenate 实现数组的拼接
a=np.array([[1,2,3],[4,5,6]])
print(a)
b=np.array([['a','b','c'],['d','e','f']])
print(b)
print(np.concatenate([a,b]))
print('垂直方向拼接 相当于 vstack')
print(np.concatenate([a,b],axis=0))
print('水平方向拼接 相当于 hstack')
print(np.concatenate([a,b],axis=1))
```

```python
#vstack 与 hstack 实现数组的拼接
print('x 轴方向及垂直堆叠')
print(np.vstack([a,b]))
print('y 轴方向及水平堆叠')
print(np.hstack([a,b]))
```

axis=0 可以使用 vstack 替换

axis=1 可以使用 hstack 替换

axis=2 可以使用 dstack 替换

```python
#三维数组的拼接
aa=np.arange(1,37).reshape(3,4,3)
print(aa)
print('-'*30)

bb=np.arange(101,137).reshape(3,4,3)
print(bb)
print('-'*30)

print('axis=0'*10)
print(np.concatenate((aa,bb),axis=0)) #6 4 3
print('axis=1'*10)
print(np.concatenate((aa,bb),axis=1)) #3,8,3
print('axis=2'*10)
print(np.concatenate((aa,bb),axis=2)) #3,4,6
```

```python
数组的分隔：numpy.split(ary, indices_or_sections, axis)
水平分隔：hsplit
垂直分隔：vsplit
```

```python
#split 分隔一维数组
x=np.arange(1,9)
y = np.split(x,4)
print(y)
print(y[2])
```

```python
z=np.split(x,[3,5])
z
```

```python
#水平分隔：hsplit
grid=np.arange(16).reshape(4,4)
a,b=np.hsplit(grid,2)

print(grid)
print()
print(a)
print()
print(b)
```

```python
grid=np.arange(16).reshape(4,4)
a,b=np.vsplit(grid,[3])

print(grid)
print()
print(a)
print()
print(b)
print()


print('三部分'*10)
a,b,c=np.vsplit(grid,[1,3])
print(a)
print()
print(b)
print()
print(c)
```

```python
#transpose 进行转置
#二维转置
a=np.arange(1,13).reshape(2,6)
print('原数组 a')
print(a)
print('转置后的数组')
print(a.transpose())
#多维数组转置
aaa=np.arange(1,37).reshape(1,3,3,4)
#将 1,3,3,4 转换为 3,3,4,1
print(np.transpose(aaa,[1,2,3,0]).shape)
```
